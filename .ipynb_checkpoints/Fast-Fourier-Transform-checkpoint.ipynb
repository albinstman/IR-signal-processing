```python
import numpy as np
from scipy import fftpack
import csv
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
%config InlineBackend.figure_format = 'svg'
```

```python
with open('cewFile0.csv', newline='') as csvfile:
    datalist = [row[0].split(",")[:-1] for row in list(csv.reader(csvfile, delimiter=' ', quotechar='|'))][2:]
```

```python
first = datalist[5930:6094]
fx = [float(row[0]) - float(first[0][0]) for row in first]
fy = [float(row[1]) for row in first]
second = datalist[9661:9825]
sx = [float(row[0]) - float(second[0][0]) for row in second]
sy = [float(row[1]) for row in second]
```

```python
fxn = np.asarray(fx)
fyn = np.asarray(fy)
sxn = np.asarray(sx)
syn = np.asarray(sy)
```

```python
fyn_fft = fftpack.fft(fyn)
syn_fft = fftpack.fft(syn)
```

```python
f_sample_freq = fftpack.fftfreq(fyn.size, d = 0.00036621)
s_sample_freq = fftpack.fftfreq(syn.size, d = 0.00036621)
```

```python
plt.plot(f_sample_freq, np.abs(fyn_fft), 'g')
plt.plot(s_sample_freq, np.abs(syn_fft), 'b')
plt.ylabel('Amplitude')
plt.show()
```

```python
plt.plot(f_sample_freq, np.angle(fyn_fft), 'g')
plt.plot(s_sample_freq, np.angle(syn_fft), 'b')
plt.ylabel('phase-shift')
plt.show()
```

```python
scale_factor = 1.02
```

```python
plt.plot(f_sample_freq/scale_factor, np.abs(fyn_fft), 'g')
plt.plot(s_sample_freq, np.abs(syn_fft), 'b')
plt.ylabel('Amplitude')
plt.show()
```

```python
plt.plot(f_sample_freq/scale_factor, np.angle(fyn_fft), 'g')
plt.plot(s_sample_freq, np.angle(syn_fft), 'b')
plt.ylabel('phase-shift')
plt.show()
```
